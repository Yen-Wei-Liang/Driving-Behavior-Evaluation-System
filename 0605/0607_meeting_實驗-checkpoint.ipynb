```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import time
import vomm

# https://github.com/rpgomez/vomm
```

```python
###########################################################################
# 步驟一：
# 取資料集(取樣頻率6)
###########################################################################
```

```python
Train_Data = pd.read_csv("Train_Data.csv")
Test_Data = pd.read_csv("Test_Data.csv")
Train_Data = Train_Data[::5]
Test_Data = Test_Data[::5]
```

```python
###########################################################################
# 步驟二：
# 提取訓練輸入序列，並做模型訓練
# 此步驟最花時間，故加入執行時間紀錄
###########################################################################
```

```python
start_time = time.time()


Go_Straight = Train_Data[Train_Data['Action'] == 'Go straight']['Base Action'].values
Go_Straight = [int(x) for x in Go_Straight]
Go_Model = vomm.ppm()
Go_Model.fit(Go_Straight, d=90, alphabet_size=4)



Idle = Train_Data[Train_Data['Action'] == 'Idle']['Base Action'].values
Idle = [int(x) for x in Idle]
Idle_Model = vomm.ppm()
Idle_Model.fit(Idle, d=90, alphabet_size=4)


Turn_Left = Train_Data[Train_Data['Action'] == 'Turn left']['Base Action'].values
Turn_Left = [int(x) for x in Turn_Left]
Left_Model = vomm.ppm()
Left_Model.fit(Turn_Left, d=90, alphabet_size=4)


Turn_Right = Train_Data[Train_Data['Action'] == 'Turn right']['Base Action'].values
Turn_Right = [int(x) for x in Turn_Right]
Right_Model = vomm.ppm()
Right_Model.fit(Turn_Right, d=90, alphabet_size=4)


Two = Train_Data[Train_Data['Action'] == 'Two-Stage Left']['Base Action'].values
Two = [int(x) for x in Two]
Two_Model = vomm.ppm()
Two_Model.fit(Two, d=90, alphabet_size=4)


U = Train_Data[Train_Data['Action'] == 'U-turn']['Base Action'].values
U = [int(x) for x in U]
U_Model = vomm.ppm()
U_Model.fit(U, d=90, alphabet_size=4)



# 計算執行時間
end_time = time.time()
execution_time = end_time - start_time
hours = int(execution_time // 3600)
minutes = int((execution_time % 3600) // 60)
seconds = int(execution_time % 60)
print(f"訓練模型所花費時間：{hours}小時{minutes}分鐘{seconds}秒")
```

```python
###########################################################################
# 步驟三：
# 提取子序列(參考時間步長 L )做測試集預測
###########################################################################
```

```python
start_time = time.time()



sequence = ''.join(Test_Data['Base Action'].astype(str))

def extract_subsequences(sequence, k):
    subsequences = []
    for i in range(len(sequence)):
        start_index = max(0, i - (k-1))  # k-1考量前k-1步(加當前則為k步)
        subsequence = sequence[start_index:i + 1]  # 提取子序列
        subsequences.append(subsequence)
    return subsequences

subsequence_6  = extract_subsequences(sequence, 6)   # 參考時間步長 L=6   (約為1秒)
subsequence_12 = extract_subsequences(sequence, 12)  # 參考時間步長 L=12  (約為2秒)
subsequence_18 = extract_subsequences(sequence, 18)  # 參考時間步長 L=18  (約為3秒)
subsequence_24 = extract_subsequences(sequence, 24)  # 參考時間步長 L=24  (約為4秒)
subsequence_30 = extract_subsequences(sequence, 30)  # 參考時間步長 L=30  (約為5秒)


# 開始預測
def predict(subsequence):
    predictions=[]
    for data in subsequence:
        Max_Score = float('-inf')
        Selected_Model = None
        data = [int(x) for x in data]
        Idle_Model_Score = Idle_Model.logpdf(data)
        Go_Model_Score = Go_Model.logpdf(data)
        Left_Model_Score = Left_Model.logpdf(data)
        Right_Model_Score = Right_Model.logpdf(data)
        Two_Model_Score = Two_Model.logpdf(data)
        U_Model_Score = U_Model.logpdf(data)
        Score_Models = [Idle_Model_Score, Go_Model_Score, Left_Model_Score, Right_Model_Score, Two_Model_Score, U_Model_Score,]    
        Models = ['Idle', 'Go Straight','Turn Left', 'Turn Right', 'Two-Stage Left', 'U-turn']    
        for Score, model in zip(Score_Models, Models):
            if Score > Max_Score:
                Max_Score = Score
                Selected_Model = model
        predictions.append(Selected_Model)
    return predictions

predictions_6  = predict(subsequence_6)
predictions_12  = predict(subsequence_12)
predictions_18  = predict(subsequence_18) 
predictions_24 = predict(subsequence_24)
predictions_30 = predict(subsequence_30)


# 將預測結果加入 Test Data (方便觀察)

Test_Data['length = 6']   = predictions_6
Test_Data['length = 12']  = predictions_12
Test_Data['length = 18']  = predictions_18
Test_Data['length = 24']  = predictions_24
Test_Data['length = 30']  = predictions_30




# 計算執行時間
end_time = time.time()
execution_time = end_time - start_time
hours = int(execution_time // 3600)
minutes = int((execution_time % 3600) // 60)
seconds = int(execution_time % 60)
print(f"預測資料集所花費時間：{hours}小時{minutes}分鐘{seconds}秒")
```

```python
def filter_actions(data):
    from statistics import mode
    filtered_data = []
    previous_action = None
    for i, action in enumerate(data):
        if action != previous_action:
            window = data[max(0, i-12):i+12]  # 取下60筆資料
            if mode(window) == action:  # 判斷眾數是否與當前資料相同
                pass 
            else:
                if mode(window + data[i-12:i+12]) != action:  # 判斷眾數是否與前一個點的資訊相同    
                    action = mode(window)  # 第一種情況，將當前點更改為眾數
                else:
                    action = previous_action  # 第二種情況，將當前點更改為前一個點的值
        filtered_data.append(action)
        previous_action = action
    return filtered_data


Test_Data['Filtered length = 6'] = filter_actions(Test_Data['length = 6'])
Test_Data['Filtered length = 12'] = filter_actions(Test_Data['length = 12'])
Test_Data['Filtered length = 18'] = filter_actions(Test_Data['length = 18'])
Test_Data['Filtered length = 24'] = filter_actions(Test_Data['length = 24'])
Test_Data['Filtered length = 30'] = filter_actions(Test_Data['length = 30'])

# 先去除未標記資料
Test_Data_Predict = Test_Data[['X-axis Angular Velocity', 'Y-axis Angular Velocity', 'Z-axis Angular Velocity', 'X-axis Acceleration', 'Y-axis Acceleration', 'Z-axis Acceleration', 'X-axis Angle', 'Y-axis Angle', 'Z-axis Angle', 'Base Action', 'Action', 'length = 6', 'length = 12', 'length = 18', 'length = 24', 'length = 30', 'Filtered length = 6', 'Filtered length = 12', 'Filtered length = 18', 'Filtered length = 24', 'Filtered length = 30']].dropna()
column_names = ['length = 6', 'length = 12', 'length = 18', 'length = 24', 'length = 30', 'Filtered length = 6', 'Filtered length = 12', 'Filtered length = 18', 'Filtered length = 24', 'Filtered length = 30']
match_percentages = []

for column in column_names:
    count = 0
    for x in range(len(Test_Data_Predict)):
        if Test_Data_Predict['Action'].iloc[x] == Test_Data_Predict[column].iloc[x]:
            count += 1
    match_percentage = (count / len(Test_Data_Predict)) * 100
    match_percentages.append(match_percentage)

result_df = pd.DataFrame({'Prediction': column_names, 'Match Percentage': match_percentages})
print(result_df)


# 存下來方便觀察(當前資料夾)
# Test_Data.to_csv("Test_Data_nontransition_Filtered_L30.csv")
```

```python
def filter_actions(data):
    from statistics import mode
    filtered_data = []
    previous_action = None
    for i, action in enumerate(data):
        if action != previous_action:
            window = data[max(0, i-12):i+12]  # 取下60筆資料
            if mode(window) == action:  # 判斷眾數是否與當前資料相同
                pass 
            else:
                if mode(window + data[i-12:i+12]) != action:  # 判斷眾數是否與前一個點的資訊相同    
                    action = mode(window)  # 第一種情況，將當前點更改為眾數
                else:
                    action = previous_action  # 第二種情況，將當前點更改為前一個點的值
        filtered_data.append(action)
        previous_action = action
    return filtered_data


Test_Data['Filtered length = 10'] = filter_actions(Test_Data['length = 10'])
Test_Data['Filtered length = 20'] = filter_actions(Test_Data['length = 20'])
Test_Data['Filtered length = 30'] = filter_actions(Test_Data['length = 30'])
Test_Data['Filtered length = 40'] = filter_actions(Test_Data['length = 40'])
Test_Data['Filtered length = 50'] = filter_actions(Test_Data['length = 50'])

# 先去除未標記資料
Test_Data_Predict = Test_Data[['X-axis Angular Velocity', 'Y-axis Angular Velocity', 'Z-axis Angular Velocity', 'X-axis Acceleration', 'Y-axis Acceleration', 'Z-axis Acceleration', 'X-axis Angle', 'Y-axis Angle', 'Z-axis Angle', 'Base Action', 'Action', 'length = 10', 'length = 20', 'length = 30', 'length = 40', 'length = 50', 'Filtered length = 10', 'Filtered length = 20', 'Filtered length = 30', 'Filtered length = 40', 'Filtered length = 50']].dropna()
column_names = ['length = 10', 'length = 20', 'length = 30', 'length = 40', 'length = 50', 'Filtered length = 10', 'Filtered length = 20', 'Filtered length = 30', 'Filtered length = 40', 'Filtered length = 50']
match_percentages = []

for column in column_names:
    count = 0
    for x in range(len(Test_Data_Predict)):
        if Test_Data_Predict['Action'].iloc[x] == Test_Data_Predict[column].iloc[x]:
            count += 1
    match_percentage = (count / len(Test_Data_Predict)) * 100
    match_percentages.append(match_percentage)

result_df = pd.DataFrame({'Prediction': column_names, 'Match Percentage': match_percentages})
print(result_df)
```

```python
Train_Data = pd.read_csv("Train_Data.csv")
Test_Data = pd.read_csv("Test_Data.csv")
```

```python
###########################################################################
# 步驟二：
# 提取訓練輸入序列，並做模型訓練
# 此步驟最花時間，故加入執行時間紀錄
###########################################################################
```

```python
start_time = time.time()


Go_Straight = Train_Data[Train_Data['Action'] == 'Go straight']['Base Action'].values
Go_Straight = [int(x) for x in Go_Straight]
Go_Model = vomm.ppm()
Go_Model.fit(Go_Straight, d=90, alphabet_size=4)



Idle = Train_Data[Train_Data['Action'] == 'Idle']['Base Action'].values
Idle = [int(x) for x in Idle]
Idle_Model = vomm.ppm()
Idle_Model.fit(Idle, d=90, alphabet_size=4)


Turn_Left = Train_Data[Train_Data['Action'] == 'Turn left']['Base Action'].values
Turn_Left = [int(x) for x in Turn_Left]
Left_Model = vomm.ppm()
Left_Model.fit(Turn_Left, d=90, alphabet_size=4)


Turn_Right = Train_Data[Train_Data['Action'] == 'Turn right']['Base Action'].values
Turn_Right = [int(x) for x in Turn_Right]
Right_Model = vomm.ppm()
Right_Model.fit(Turn_Right, d=90, alphabet_size=4)


Two = Train_Data[Train_Data['Action'] == 'Two-Stage Left']['Base Action'].values
Two = [int(x) for x in Two]
Two_Model = vomm.ppm()
Two_Model.fit(Two, d=90, alphabet_size=4)


U = Train_Data[Train_Data['Action'] == 'U-turn']['Base Action'].values
U = [int(x) for x in U]
U_Model = vomm.ppm()
U_Model.fit(U, d=90, alphabet_size=4)



# 計算執行時間
end_time = time.time()
execution_time = end_time - start_time
hours = int(execution_time // 3600)
minutes = int((execution_time % 3600) // 60)
seconds = int(execution_time % 60)
print(f"訓練模型所花費時間：{hours}小時{minutes}分鐘{seconds}秒")
```

```python
###########################################################################
# 步驟三：
# 提取子序列(參考時間步長 L )做測試集預測
###########################################################################
```

```python
start_time = time.time()



sequence = ''.join(Test_Data['Base Action'].astype(str))

def extract_subsequences(sequence, k):
    subsequences = []
    for i in range(len(sequence)):
        start_index = max(0, i - (k-1))  # k-1考量前k-1步(加當前則為k步)
        subsequence = sequence[start_index:i + 1]  # 提取子序列
        subsequences.append(subsequence)
    return subsequences

subsequence_30  = extract_subsequences(sequence, 30)   # 參考時間步長 L=30   (約為1秒)
subsequence_60  = extract_subsequences(sequence, 60)   # 參考時間步長 L=60   (約為2秒)
subsequence_90  = extract_subsequences(sequence, 90)   # 參考時間步長 L=90   (約為3秒)
subsequence_120 = extract_subsequences(sequence, 120)  # 參考時間步長 L=120  (約為4秒)
subsequence_150 = extract_subsequences(sequence, 150)  # 參考時間步長 L=150  (約為5秒)


# 開始預測
def predict(subsequence):
    predictions=[]
    for data in subsequence:
        Max_Score = float('-inf')
        Selected_Model = None
        data = [int(x) for x in data]
        Idle_Model_Score = Idle_Model.logpdf(data)
        Go_Model_Score = Go_Model.logpdf(data)
        Left_Model_Score = Left_Model.logpdf(data)
        Right_Model_Score = Right_Model.logpdf(data)
        Two_Model_Score = Two_Model.logpdf(data)
        U_Model_Score = U_Model.logpdf(data)
        Score_Models = [Idle_Model_Score, Go_Model_Score, Left_Model_Score, Right_Model_Score, Two_Model_Score, U_Model_Score,]    
        Models = ['Idle', 'Go Straight','Turn Left', 'Turn Right', 'Two-Stage Left', 'U-turn']    
        for Score, model in zip(Score_Models, Models):
            if Score > Max_Score:
                Max_Score = Score
                Selected_Model = model
        predictions.append(Selected_Model)
    return predictions

predictions_30   = predict(subsequence_30)
predictions_60   = predict(subsequence_60)
predictions_90   = predict(subsequence_90) 
predictions_120  = predict(subsequence_120)
predictions_150  = predict(subsequence_150)


# 將預測結果加入 Test Data (方便觀察)

Test_Data['length = 30']   = predictions_30
Test_Data['length = 60']   = predictions_60
Test_Data['length = 90']   = predictions_90
Test_Data['length = 120']  = predictions_120
Test_Data['length = 150']  = predictions_150




# 計算執行時間
end_time = time.time()
execution_time = end_time - start_time
hours = int(execution_time // 3600)
minutes = int((execution_time % 3600) // 60)
seconds = int(execution_time % 60)
print(f"預測資料集所花費時間：{hours}小時{minutes}分鐘{seconds}秒")
```

```python
def filter_actions(data):
    from statistics import mode
    filtered_data = []
    previous_action = None
    for i, action in enumerate(data):
        if action != previous_action:
            window = data[max(0, i-12):i+12]  # 取下60筆資料
            if mode(window) == action:  # 判斷眾數是否與當前資料相同
                pass 
            else:
                if mode(window + data[i-12:i+12]) != action:  # 判斷眾數是否與前一個點的資訊相同    
                    action = mode(window)  # 第一種情況，將當前點更改為眾數
                else:
                    action = previous_action  # 第二種情況，將當前點更改為前一個點的值
        filtered_data.append(action)
        previous_action = action
    return filtered_data


Test_Data['Filtered length = 30']  = filter_actions(Test_Data['length = 30'])
Test_Data['Filtered length = 60']  = filter_actions(Test_Data['length = 60'])
Test_Data['Filtered length = 90']  = filter_actions(Test_Data['length = 90'])
Test_Data['Filtered length = 120'] = filter_actions(Test_Data['length = 120'])
Test_Data['Filtered length = 150'] = filter_actions(Test_Data['length = 150'])

# 先去除未標記資料
Test_Data_Predict = Test_Data[['X-axis Angular Velocity', 'Y-axis Angular Velocity', 'Z-axis Angular Velocity', 'X-axis Acceleration', 'Y-axis Acceleration', 'Z-axis Acceleration', 'X-axis Angle', 'Y-axis Angle', 'Z-axis Angle', 'Base Action', 'Action', 'length = 30', 'length = 60', 'length = 90', 'length = 120', 'length = 150', 'Filtered length = 30', 'Filtered length = 60', 'Filtered length = 90', 'Filtered length = 120', 'Filtered length = 150']].dropna()
column_names = ['length = 30', 'length = 60', 'length = 90', 'length = 120', 'length = 150', 'Filtered length = 30', 'Filtered length = 60', 'Filtered length = 90', 'Filtered length = 120', 'Filtered length = 150']
match_percentages = []

for column in column_names:
    count = 0
    for x in range(len(Test_Data_Predict)):
        if Test_Data_Predict['Action'].iloc[x] == Test_Data_Predict[column].iloc[x]:
            count += 1
    match_percentage = (count / len(Test_Data_Predict)) * 100
    match_percentages.append(match_percentage)

result_df = pd.DataFrame({'Prediction': column_names, 'Match Percentage': match_percentages})
print(result_df)
```
