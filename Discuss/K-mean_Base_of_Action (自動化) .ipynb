# Case 1 : Normalizatoin + Kmean

```python
#############################
# 讀取檔案
#############################
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
Data_Set = pd.read_csv(r"C:\Users\William\Desktop\20230420\分5群\Action.csv")
Data_Set
```

```python
#############################
# 只取9軸特徵
#############################
Data_Set_label = Data_Set['Action']
Data_Set = Data_Set[['X-axis Angular Velocity','Y-axis Angular Velocity','Z-axis Angular Velocity','X-axis Acceleration','Y-axis Acceleration','Z-axis Acceleration','X-axis Angle', 'Y-axis Angle', 'Z-axis Angle']]


Data_Set
```

```python
#############################
# 正規化
#############################

from sklearn.preprocessing import MinMaxScaler
import pandas as pd

# 創建MinMaxScaler對象
scaler = MinMaxScaler()

# 使用MinMaxScaler對DataFrame進行最小-最大標準化
Data_Set = pd.DataFrame(scaler.fit_transform(Data_Set), columns=Data_Set.columns)

Data_Set
```

```python
#############################
# k-mean分群 5
#############################

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA


# 建立 KMeans 模型並進行分群
kmeans = KMeans(n_clusters=5)
kmeans.fit(Data_Set)
labels = kmeans.labels_


Data_Set['Base Action'] = labels
Data_Set['Action'] = Data_Set_label
```

```python
import pandas as pd

# 建立空的 DataFrame
result = pd.DataFrame(columns=['Action', 'Base Action', 'Number', 'Rate'])

for x in sorted(Data_Set['Action'].unique()):
    for y in sorted(Data_Set['Base Action'].unique()):
        total = len(Data_Set[(Data_Set['Action']==x)])
        Base_total = len(Data_Set[(Data_Set['Action']==x) & (Data_Set['Base Action']==y)])
        rate = round((Base_total/total)*100, 2)
        result.loc[len(result)] = [x, y, Base_total, rate]

        
centers = pd.DataFrame(kmeans.cluster_centers_)
centers = centers.rename(index={0: 'Group 0', 1: 'Group 1', 2: 'Group 2', 3: 'Group 3', 4: 'Group 4'})
centers.columns =['X-axis Angular Velocity','Y-axis Angular Velocity','Z-axis Angular Velocity','X-axis Acceleration','Y-axis Acceleration','Z-axis Acceleration','X-axis Angle', 'Y-axis Angle', 'Z-axis Angle' ]
        
    

#顯示結果
centers.to_csv(r"C:\Users\William\Desktop\20230420\分5群\Case_1_centers.csv")
result.to_csv(r"C:\Users\William\Desktop\20230420\分5群\Case_1_rate.csv")
Data_Set.to_csv(r"C:\Users\William\Desktop\20230420\分5群\Case_1.csv")
```
